```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import scipy.stats as stats
from scipy.stats import cauchy
from scipy.optimize import minimize
from sklearn import linear_model
from dateutil.relativedelta import relativedelta
import warnings
warnings.filterwarnings('ignore')
```

```python
def process_econ_data(econ):
    Name_List = ["ECON", "GDP", "DGR", "CDS", "CPI", "DEF", "GB1", "GB2", "GB5", "GB10", "GB30"]
    File_List = ["Countries.pkl", "GDP.pkl", "DGR.pkl", "CDS.pkl", "CPI.pkl", "DEF.pkl",
                "GOV1Y.pkl", "GOV2Y.pkl", "GOV5Y.pkl", "GOV10Y.pkl", "GOV30Y.pkl"]
    for i in range(len(Name_List)):
        with open(File_List[i], "rb") as file:
            globals()[Name_List[i]] = pickle.load(file)
    Data_List = [GDP, DGR, CDS, CPI, DEF, GB1, GB2, GB5, GB10, GB30]

    code = ECON.loc[econ, "abbr"]
    date = code + "_date"
    for i in range(len(Data_List)):
        globals()[code + Name_List[i+1]] = Data_List[i].loc[:, [date, code]]
        globals()[code + Name_List[i+1]][date] = pd.to_datetime(globals()[code + Name_List[i+1]][date])
        globals()[code + Name_List[i+1]].set_index(date, inplace = True)

    globals()[code + "GDP"]["Growth"] = globals()[code + "GDP"][code].pct_change()
    globals()[code + "GDP"]["5YMA"] = globals()[code + "GDP"]['Growth'].rolling(window=5).apply(lambda x: (x * 0.2).sum(),
                                                                                                raw=True)
    globals()[code + "GDP"] = globals()[code + "GDP"].iloc[5:]
    globals()[code + "DEF_Y"] = globals()[code + "DEF"].resample("Y").sum()
    globals()[code + "GB1_Y"] = globals()[code + "GB1"].resample("Y").mean()

    globals()[code + "DEF_Y"]["Surplus"] = globals()[code + "DEF_Y"].loc[globals()[code + "DEF_Y"][code] > 0, code]
    globals()[code + "DEF_Y"]["Surplus"].fillna(0, inplace = True)
    globals()[code + "DEF_Y"]["Deficit"] = globals()[code + "DEF_Y"].loc[globals()[code + "DEF_Y"][code] <= 0, code]
    globals()[code + "DEF_Y"]["Deficit"].fillna(0, inplace = True)

    mean_g = np.mean(globals()[code + "GDP"]["Growth"])

    def implied_pd_from_cds_price(cds_price,rr,rf):
        implied_pd =  1- np.exp((-cds_price)/(1-rr))
        return implied_pd
    def cauchy_log_likelihood(params, data):
        loc, scale = params
        return -np.sum(np.log(cauchy.pdf(data, loc, scale)))

    x0 = [mean_g, 0.05]
    data = np.array([x for x in globals()[code + "GDP"]["Growth"] if not np.isnan(x) and not np.isinf(x)])
    constraints = ({"type": "ineq", "fun": lambda x: x[1]})
    result = minimize(cauchy_log_likelihood, x0, args = (data,), constraints = constraints)
    fitted_location_parameter, fitted_scale_parameter = result.x


    merge = pd.merge(globals()[code + "GDP"], globals()[code + "DGR"], on = date, how = "outer")
    merge = pd.merge(merge, globals()[code + "DEF_Y"], on = date, how = "outer")
    merge = pd.merge(merge, globals()[code + "GB1_Y"], on = date, how = "outer")
    merge.columns = ["gdp", "Growth", "5YMA", "debt_to_gdp", "Fiscal", "Surplus", "Deficit", "1YGB"]
    df = pd.DataFrame({
        "debt_to_gdp" : merge["debt_to_gdp"] / 100,
        "debt" : merge["debt_to_gdp"] / 100 * merge["gdp"],
        "MPS" : (merge["Fiscal"] / merge["gdp"]),
        "gdp" : merge["gdp"],
        "gt" : merge["Growth"],
    })
    df["debt_pct_change"] = df["debt"].pct_change() * 100
    df.dropna(inplace = True)
    df

    data_debt = df["debt_pct_change"].dropna()
    data_debt = np.array([x for x in data_debt if not np.isnan(x) and not np.isinf(x)])
    x0_debt = [np.mean(data_debt), 0.05]
    constraints_debt = ({"type": "ineq", "fun": lambda x: x[1]})
    result_debt = minimize(cauchy_log_likelihood, x0_debt, args = (data_debt,), constraints = constraints_debt)
    fitted_location_parameter_debt, fitted_scale_parameter_debt = result_debt.x


    np.random.seed(1234)
    samples_gdp_growth = cauchy.rvs(loc = fitted_location_parameter, scale = fitted_scale_parameter, size = (5,10000))
    samples_debt_pct_change = cauchy.rvs(loc = fitted_location_parameter_debt,
                                        scale = fitted_scale_parameter_debt,
                                        size = (5,10000))
    samples_gdp_growth_real = np.prod(1 + samples_gdp_growth / 100, axis=0) * df["gdp"][-1]
    samples_debt_pct_change_real = np.prod(1 + samples_debt_pct_change / 100, axis = 0)* df["debt"][-1]
    # Dividing the values from samples_debt_pct_change by samples_gdp_growth
    resultant_values = samples_debt_pct_change_real / samples_gdp_growth_real

    critical_min_gdp_growth_rate = fitted_location_parameter
    x = np.linspace(-1, 1, 100)
    g_cdf = cauchy.cdf(x, loc = critical_min_gdp_growth_rate, scale = np.std(df["gt"]))
    df["threshold"] = df.apply(lambda row: row["debt_to_gdp"] + min(row["gt"], critical_min_gdp_growth_rate), axis = 1)

    closest_index = np.argmin(np.abs(x - critical_min_gdp_growth_rate))
    fitted_mean_cdf_value = g_cdf[closest_index]
    print("Prob of growth rate being lower or equal the critical ", fitted_mean_cdf_value *100,"%" )
    ## growth mean implies the fiscal space meaning capacity/expectation that gov can handle above its debt-to-gdp ratio

    # Finding the value where CDF is 80%
    sorted_data_dtg = np.sort(resultant_values)
    yvals = np.arange(1, len(sorted_data_dtg) + 1) / float(len(sorted_data_dtg))
    globals()[code + "CDS"] = globals()[code + "CDS"].dropna()
    rr = 0.4
    '''
    rr = [0.799, 0.867, 0.419, 0.369, 0.885,
          0.748, 0.798, 0.716, 0.656, 0.921,
          0.901, 0.313, 0.648, 0.861, 0.461,
          0.843, 0.775, 0.854, 0.810, 0.213]
    '''
    ipd = 1-np.exp((-globals()[code + "CDS"].iloc[-1,0]/100)/(1-rr))
    Debt_GDP_cutoff = np.percentile(resultant_values, (1-ipd)*100)

    Debt_GDP_cutoff

    with open("DGR30.pkl", "rb") as file:
        DGR30 = pickle.load(file)
    country = ECON["country"].tolist()
    wbdata = ["GHGE", "RENT", "RNEW"]
    wbdict = {"Republic of Ireland":"Ireland",
              "Slovakia":"Slovak Republic",
              "South Korea":"Korea, Rep.",
              "Vietnam":"Viet Nam"}
    country = [wbdict.get(country, country) for country in country]
    ECON.loc[:, "country"] = country
    name = ECON.loc[econ, "country"]
    for i in wbdata:
        file_name = i + ".csv"
        globals()[i] = pd.read_csv(file_name, skiprows = [0, 1, 2, 3])
        globals()[i] = globals()[i][globals()[i]["Country Name"].isin(country)].reset_index(drop = True)
        globals()[i] = globals()[i].dropna(axis = 1)
        globals()[i] = globals()[i].drop(["Country Code", "Indicator Name", "Indicator Code"], axis = 1)
        globals()[i] = globals()[i].T
        globals()[i].columns = globals()[i].iloc[0]
        globals()[i] = globals()[i].iloc[1:]

    econ_dgr = DGR30.loc[:, [code, date]].dropna()
    econ_dgr[date] = [dt.strftime("%Y") for dt in econ_dgr[date]]
    econ_dgr = econ_dgr.rename(columns = {date:"index"})
    econ_ghg = GHGE.loc[:, name].reset_index()
    econ_rnt = RENT.loc[:, name].reset_index()
    econ_new = RNEW.loc[:, name].reset_index()
    reg_df = pd.merge(econ_dgr, econ_ghg, on = "index", how = "inner")
    reg_df = pd.merge(reg_df, econ_rnt, on = "index", how = "inner")
    reg_df = pd.merge(reg_df, econ_new, on = "index", how = "inner")
    reg_df = reg_df.rename(columns = {code:"dgr",
                                      "index":"year",
                                      name+"_x":"ghg",
                                      name+"_y":"rent",
                                      name:"renew"})
    reg_df["lagdgr"] = reg_df["dgr"].shift(1)
    reg_df["lagghg"] = reg_df["ghg"].shift(1)
    reg_df["lagrent"] = reg_df["rent"].shift(1)
    reg_df["lagnew"] = reg_df["renew"].shift(1)
    reg_df = reg_df.dropna().reset_index(drop = True)
    reg_df["difdgr"] = reg_df["dgr"] - reg_df["lagdgr"]
    reg_df["difghg"] = reg_df["ghg"] - reg_df["lagghg"]
    reg_df["difrent"] = reg_df["rent"] - reg_df["lagrent"]
    reg_df["difnew"] = reg_df["renew"] - reg_df["lagnew"]
    reg_df["lagdgr"] = reg_df["difdgr"].shift(1)
    reg_df["lagghg"] = reg_df["difghg"].shift(1)
    reg_df["lagrent"] = reg_df["difrent"].shift(1)
    reg_df["lagnew"] = reg_df["difnew"].shift(1)
    reg_df = reg_df.dropna().reset_index(drop = True)

    X = reg_df[["lagdgr", "difghg", "difrent", "difnew"]]
    y = reg_df[["difdgr"]]
    regr = linear_model.LinearRegression()
    regr.fit(X, y)
    regint = regr.intercept_[0]
    regmul = regr.coef_[0][0]
    regcoef = regr.coef_[0][1:].tolist()

    X = reg_df[["lagghg"]]
    y = reg_df[["difghg"]]
    regr = linear_model.LinearRegression()
    regr.fit(X, y)
    ghgint, ghgcoef = regr.intercept_[0], regr.coef_[0][0]

    X = reg_df[["lagrent"]]
    y = reg_df[["difrent"]]
    regr = linear_model.LinearRegression()
    regr.fit(X, y)
    regr.intercept_, regr.coef_
    rentint, rentcoef = regr.intercept_[0], regr.coef_[0][0]

    X = reg_df[["lagnew"]]
    y = reg_df[["difnew"]]
    regr = linear_model.LinearRegression()
    regr.fit(X, y)
    newint, newcoef = regr.intercept_[0], regr.coef_[0][0]

    df['growth'] = df['gdp'].pct_change()*100
    df.dropna(inplace=True)

    def sample(gdp_growth_adjustment, years_size):
        # Adjusting the gdp['growth']
        data_85 = df['growth'] - gdp_growth_adjustment * 100 /27                #in percentage points and divide by 27 years(by 2050)
        data_85 = np.array([x for x in data_85 if not np.isnan(x) and not np.isinf(x)])

        def cauchy_log_likelihood(params, data):
            loc, scale = params
            return -np.sum(np.log(cauchy.pdf(data, loc, scale)))

        x0 = [np.mean(data_85), 0.05]
        constraints = ({'type': 'ineq', 'fun': lambda x: x[1]})
        result = minimize(cauchy_log_likelihood, x0, args=(data_85,), constraints=constraints)
        new_fitted_location_parameter, new_fitted_scale_parameter = result.x

        df['debt_pct_change'] = df['debt'].pct_change() * 100
        data_debt = df['debt_pct_change'].dropna()
        data_debt = np.array([x for x in data_debt if not np.isnan(x) and not np.isinf(x)])
        def cauchy_log_likelihood(params, data):
            loc, scale = params
            return -np.sum(np.log(cauchy.pdf(data, loc, scale)))

        x0 = [np.mean(data_85), 0.05]
        constraints = ({'type': 'ineq', 'fun': lambda x: x[1]})
        result = minimize(cauchy_log_likelihood, x0, args=(data_85,), constraints=constraints)
        new_fitted_location_parameter, new_fitted_scale_parameter = result.x

        #New, Transitional risk on debt with our simplified model
        transitional_risk = pd.read_csv("Transitional Risk.csv")
        debt_penalty = transitional_risk.iloc[econ-1,4]

        df['debt_pct_change'] = df['debt'].pct_change() * 100 + debt_penalty
        data_debt = df['debt_pct_change'].dropna()
        data_debt = np.array([x for x in data_debt if not np.isnan(x) and not np.isinf(x)])

        def cauchy_log_likelihood_debt(params, data):
            loc, scale = params
            return -np.sum(np.log(cauchy.pdf(data, loc, scale)))

        x0_debt = [np.mean(data_debt), 0.05]
        constraints_debt = ({'type': 'ineq', 'fun': lambda x: x[1]})
        result_debt = minimize(cauchy_log_likelihood_debt, x0_debt, args=(data_debt,), constraints=constraints_debt)
        fitted_location_parameter_debt, fitted_scale_parameter_debt = result_debt.x


        # Drawing random numbers from the Cauchy distributions using the provided years size
        np.random.seed(1234)
        yearss_gdp_growth = cauchy.rvs(loc=fitted_location_parameter, scale=fitted_scale_parameter, size=(years_size, 10000))
        yearss_debt_pct_change = cauchy.rvs(loc=fitted_location_parameter_debt, scale=fitted_scale_parameter_debt, size=(years_size, 10000))
        new_yearss_gdp_growth = cauchy.rvs(loc=new_fitted_location_parameter, scale=new_fitted_scale_parameter, size=(years_size, 10000))

        yearss_gdp_growth_real = np.prod(1 + yearss_gdp_growth / 100, axis=0) * df['gdp'][-1]
        yearss_debt_pct_change_real = np.prod(1 + yearss_debt_pct_change / 100, axis=0) * df['debt'][-1]
        new_yearss_gdp_growth_real = np.prod(1 + new_yearss_gdp_growth / 100, axis=0) * df['gdp'][-1]

        resultant_values = yearss_debt_pct_change_real / yearss_gdp_growth_real
        new_resultant_values = yearss_debt_pct_change_real / new_yearss_gdp_growth_real
        return new_resultant_values

    def sample_climate(gdp_growth_adjustment, years_size):
        # Adjusting the gdp['growth']
        data_85 = df['growth'] - gdp_growth_adjustment * 100 /27                #in percentage points and divide by 27 years(by 2050)
        data_85 = np.array([x for x in data_85 if not np.isnan(x) and not np.isinf(x)])

        def cauchy_log_likelihood(params, data):
            loc, scale = params
            return -np.sum(np.log(cauchy.pdf(data, loc, scale)))

        x0 = [np.mean(data_85), 0.05]
        constraints = ({'type': 'ineq', 'fun': lambda x: x[1]})
        result = minimize(cauchy_log_likelihood, x0, args=(data_85,), constraints=constraints)
        new_fitted_location_parameter, new_fitted_scale_parameter = result.x

        #New, Transitional risk on debt with our simplified model
        transitional_risk = pd.read_csv("Transitional Risk.csv")
        debt_penalty = transitional_risk.iloc[econ-1,4] *2

        df['debt_pct_change'] = df['debt'].pct_change() * 100 + debt_penalty
        data_debt = df['debt_pct_change'].dropna()
        data_debt = np.array([x for x in data_debt if not np.isnan(x) and not np.isinf(x)])

        def cauchy_log_likelihood_debt(params, data):
            loc, scale = params
            return -np.sum(np.log(cauchy.pdf(data, loc, scale)))

        x0_debt = [np.mean(data_debt), 0.05]
        constraints_debt = ({'type': 'ineq', 'fun': lambda x: x[1]})
        result_debt = minimize(cauchy_log_likelihood_debt, x0_debt, args=(data_debt,), constraints=constraints_debt)
        fitted_location_parameter_debt, fitted_scale_parameter_debt = result_debt.x


        # Drawing random numbers from the Cauchy distributions using the provided years size
        np.random.seed(1234)
        yearss_gdp_growth = cauchy.rvs(loc=fitted_location_parameter, scale=fitted_scale_parameter, size=(years_size, 10000))
        yearss_debt_pct_change = cauchy.rvs(loc=fitted_location_parameter_debt, scale=fitted_scale_parameter_debt, size=(years_size, 10000))
        new_yearss_gdp_growth = cauchy.rvs(loc=new_fitted_location_parameter, scale=new_fitted_scale_parameter, size=(years_size, 10000))

        yearss_gdp_growth_real = np.prod(1 + yearss_gdp_growth / 100, axis=0) * df['gdp'][-1]
        yearss_debt_pct_change_real = np.prod(1 + yearss_debt_pct_change / 100, axis=0) * df['debt'][-1]
        new_yearss_gdp_growth_real = np.prod(1 + new_yearss_gdp_growth / 100, axis=0) * df['gdp'][-1]

        resultant_values = yearss_debt_pct_change_real / yearss_gdp_growth_real
        new_resultant_values = yearss_debt_pct_change_real / new_yearss_gdp_growth_real
        return new_resultant_values

    gdp_penalty = pd.read_csv("GDP_Penalty.csv")
    short_basecase = sample(0, 7)
    long_basecase = sample(0, 27)
#3x GDP penality for 10x scenarios
    short_A  = sample_climate(-gdp_penalty.iloc[econ-1,0]*10, 7)
    long_A  = sample_climate(-gdp_penalty.iloc[econ-1,0]*10, 27)
    short_B  = sample_climate(-gdp_penalty.iloc[econ-1,1]*10, 7)
    long_B  = sample_climate(-gdp_penalty.iloc[econ-1,1]*10, 27)
    short_C  = sample_climate(-gdp_penalty.iloc[econ-1,2]*10, 7)
    long_C  = sample_climate(-gdp_penalty.iloc[econ-1,2]*10, 27)
    short_D  = sample_climate(-gdp_penalty.iloc[econ-1,3]*10, 7)
    long_D  = sample_climate(-gdp_penalty.iloc[econ-1,3]*10, 27)

    # PD of short
    def PD(data):
        sorted_data = np.sort(data)
        idx = np.searchsorted(sorted_data, Debt_GDP_cutoff)
        yvals = np.arange(1, len(sorted_data) + 1) / float(len(sorted_data))
        interp_yval = np.interp(Debt_GDP_cutoff, sorted_data, yvals)
        PD = (1-interp_yval)*100
        return PD

    PD_summary = pd.DataFrame()
    PD_summary["2030"] = [PD(short_basecase),PD(short_A),PD(short_B),PD(short_C),PD(short_D)]
    PD_summary["2050"] = [PD(long_basecase),PD(long_A),PD(long_B),PD(long_C),PD(long_D)]
    PD_summary.index = ["Base Case", "1.5", "2", "2.6", "3.2"]

    # Add the country name to the PD_summary DataFrame
    PD_summary['Country'] = name

    # Print the PD_summary DataFrame
    print(PD_summary.to_markdown())

    bin_edges = np.linspace(0, 10, 201)

    # Create a histogram with horizontal orientation
    plt.hist(short_basecase,bins=bin_edges,alpha = 0.3, color='black', orientation='horizontal',label = "No penalty")

    #plt.hist(short_A, bins=bin_edges, color='red', alpha=0.5, orientation='horizontal',label = "lower 2 degrees")
    plt.hist(short_B, bins=bin_edges, color='red', alpha=0.5, orientation='horizontal',label = "2 degrees")
    #plt.hist(short_C, bins=bin_edges, color='green', alpha=0.5, orientation='horizontal',label = "2.6 degrees")
    plt.hist(short_D, bins=bin_edges, color='green', alpha=0.5, orientation='horizontal',label = "3.2 degrees")


    # Add a shaded region on the y-axis with adjusted position and size
    plt.axhspan(Debt_GDP_cutoff, 10, facecolor='red', alpha=0.1, label='default zone')

    # Customize the plot
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.xlabel('Frequency')
    plt.ylabel('Debt/GDP')
    plt.title(f'2030 Climate adjusted Debt/GDP distribution for {name}')
    plt.ylim(0,10)
    plt.axhline(y=Debt_GDP_cutoff,alpha = 0.3, color='r', linestyle='--', label=f'Debt / GDP cutoff = {Debt_GDP_cutoff:.2f}')
    plt.legend()

    # Set the y-axis limits to make the histogram start from the right side
    plt.ylim(plt.ylim()[::-1])

    # Show the plot
    plt.show()

    bin_edges = np.linspace(0, 10, 201)

    # Create a histogram with horizontal orientation
    plt.hist(long_basecase, bins=bin_edges,alpha = 0.3, color='black', orientation='horizontal',label = "No penalty")

    #plt.hist(long_A, bins=bin_edges, color='red', alpha=0.5, orientation='horizontal',label = "lower 2 degrees")

    plt.hist(long_B, bins=bin_edges, color='red', alpha=0.5, orientation='horizontal',label = "2 degrees")
    #plt.hist(long_C, bins=bin_edges, color='green', alpha=0.5, orientation='horizontal',label = "2.6 degrees")
    plt.hist(long_D, bins=bin_edges, color='green', alpha=0.5, orientation='horizontal',label = "3.2 degrees")
    plt.axhline(y=Debt_GDP_cutoff,alpha = 0.3, color='r', linestyle='--', label=f'Debt / GDP cutoff = {Debt_GDP_cutoff:.2f}')


    # Add a shaded region on the y-axis with adjusted position and size
    plt.axhspan(Debt_GDP_cutoff, 10, facecolor='red', alpha=0.1, label='default zone')

    # Customize the plot
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.xlabel('Frequency')
    plt.ylabel('Debt/GDP')
    plt.title(f'2050 Climate adjusted Debt/GDP distribution for {name}')
    plt.ylim(0,10)
    plt.legend()

    # Set the y-axis limits to make the histogram start from the right side
    plt.ylim(plt.ylim()[::-1])

    # Show the plot
    plt.show()


for econ in range(1,19):
    process_econ_data(econ)
```
